```python
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())
df['NumberOfDependents'] = df['NumberOfDependents'].fillna(df['NumberOfDependents'].median())
```

```python
df.isnull().sum()
```

```python
#Detecting outliers
```

```python
Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

lower_bound, upper_bound
```

```python
df['MonthlyIncome'].describe()
```

```python
#Cap Outliers

df['MonthlyIncome'] = df['MonthlyIncome'].clip(lower_bound,upper_bound)
```

```python
# 96 and 98 replacements
df['NumberOfTimes90DaysLate']=df['NumberOfTimes90DaysLate'].replace([96,98],10)
```

```python
df['NumberOfTimes90DaysLate'].value_counts()
```

```python
#apply scaling
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=['SeriousDlqin2yrs'])
Y = df['SeriousDlqin2yrs']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#train test split (80:20)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled, Y, 
    test_size=0.2,
    random_state=42
)
```

```python
#code train baseline logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight='balanced')
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
print(cm)

#model is not bad at finding defaulters, but misses too many good customers
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))

#shitty precision for 1
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight={0:1, 1:3})
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))

#nochange :(
```

```python
#smote - synthetic minority oversampling
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state = 42)
X_train_sm, Y_train_sm = smote.fit_resample(X_train,Y_train)
```

```python
print(Y_train.value_counts())
```

```python
print(Y_train_sm.value_counts())
```

```python
model = LogisticRegression()
model.fit(X_train_sm, Y_train_sm)
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))

#SMOTE alone did not improve recall with Logistic Regression. Model capacity matters
```

```python
y_prob = model.predict_proba(X_test)[:, 1]

y_pred_new = (y_prob > 0.3).astype(int)
```

```python
print(confusion_matrix(Y_test, y_pred_new))
print(classification_report(Y_test, y_pred_new))
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=5, random_state=42)

model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```
